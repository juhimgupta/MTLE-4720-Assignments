# Improving Data Visualization in ICU/Population Data and Marathon Runners Data

First, lets import all the tools we need to improve data visualization for these two datasets. 

```python
import scipy
from scipy.optimize import curve_fit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition,
                                                  mark_inset)
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
```

## Number of ICU Beds and Population 

Now, lets load the ICU data from the CSV excel sheet using *pd.read_csv.*

```python
icudata = pd.read_csv(r"C:\Users\guptaj\Documents\ICUdata.csv")
```

Next, let's print the data so that we can see the columns and the column names.

```python
print(icudata)
```

Since we are trying to compare the number of ICU beds to the total population in state counties, we should obtain those to columns of data and save it to another variables so that they are easier to work with! Here, we save the total population column in *x* and the number of ICU beds column in *y.*

```python
x = icudata['Total Population']
y = icudata['ICU Beds']
```

### Original Graph 

Let's state by plotting the original graph to see what we can improve. We can use *plt.scatter* to plot a scatter plot and then use *plt.title(), plt.xlabel(),* and *plt.ylabel* to put graph and axis titles. 

```python
plt.scatter(x,y,c='purple')
plt.title('Number of ICU Beds vs Population in US State Counties')
plt.xlabel('Population in County')
plt.ylabel('Number of ICU Beds')
```

### Limiting Axes to See Data Points Better

One problem with the original graph is that it is very difficult to see the data for the number of ICU beds for counties with lower population. Instead of seeing individual data points, we see a clump of purple. Generating a subplot zoomed in on the clumped up scatter is a way we can see that data better.  

Lets try to shows populations in the county up to 2,000,000 with a maximum number of ICU Beds at 750. This "box" covers most of the data points in that purple clump and gives viewers a better understanding of what the number of ICU Beds vs Population for most US counties looks like. 

We can use *x.lim* and *y.lim* to set the intervals of data we want to look at. Since we want to look at counties with populations less than or equal to 2,000,000, our interval will be (0,2000000). Since we want to look at the points where the number of ICU beds is less than or equal to 750, our interval will be (0,750). 

```python
plt.scatter(x,y,c = 'purple')
plt.xlim(0,2000000)
plt.ylim(0,750)
plt.title('Number of ICU Beds vs Population in US State Counties')
plt.xlabel('Population in County')
plt.ylabel('Number of ICU Beds')
```

### Making an Inset Graph in the Original Data

We can also put a close up of the "box" we constructed earlier inside the original graph. This is called an inset. In order to do this we use *fig, ax = plt.subplots()* to create a figure and a subplot. 

Next, plot the original graph but set *plt.scatter* to *ax* to make a large subplot. 

Then, we can position another plot, called *ax2* in the larger figure so that it is inside the the original graph, *ax.* Using the notation that (0,0) is the bottom left of the figure, we set numbers to varibles *left, bottom, width,* and *height* and put those values into *fig.add_axes* to create a new axes at that location. 

Next, we can actually plot the ICU data in that smaller plot, *ax2,* using *ax2.plot,* and following the same procedure that was used for limiting axes (using *.xlim* and *.ylim.* 

Finally, we can use *plt.title* after the *ax2.plot* to label specifically that graph. 


```python
fig, ax = plt.subplots()

ax = plt.scatter(x,y,c='purple')

plt.title('Number of ICU Beds vs Population in US State Counties')
plt.xlabel('Population in County')
plt.ylabel('Number of ICU Beds')

# These are in unitless percentages of the figure size. (0,0 is bottom left)
left, bottom, width, height = [0.5, 0.18, 0.4, 0.3]
ax2 = fig.add_axes([left, bottom, width, height])

ax2.plot(x, y,'go')
plt.xlim(0,2000000)
plt.ylim(0,750)

plt.title('Zoomed Image')
```

### More Ways to Visualize This Data 

There are lots of ways to interpret and visualize the ICU data. We could color code the scatter points according to region in the US, plot how many ICU beds there are for people aged 60+ in each state, etc. In our example, We  use the ICU data to look at how many people there are per ICU bed in states. To make this simpler, lets use only Northeastern States.

First, lets get the information we need. From the *icudata* dataframe, we first pick out the state information want and save that into a new variable. In the first line, we see that we have set a condition that from the *icudata* we want to save only those values whose *State* is equal to *New Hampshire.* 

To find the total population of New Hampshire, we can add the population of all the counties, which is stored in *Total Population,* together and save it in NHPop. To find the total number of ICU beds in New Hampshire, we can add the number of ICU beds of all the counties, which is stored in *ICU Beds,* together and save it in NHBeds. 

Next, to find the number of people per ICU bed, we can divide NHPop by NHBeds and save it in NHPopBed. 

We can do the same thing for the rest of the states! 

```python
NHdata = icudata[icudata['State'] == 'New Hampshire']
NHPop = sum(NHdata['Total Population'])
NHBeds = sum(NHdata['ICU Beds'])
NHPopBed = NHPop/NHBeds

MEdata = icudata[icudata['State'] == 'Maine']
MEPop = sum(MEdata['Total Population'])
MEBeds = sum(MEdata['ICU Beds'])
MEPopBed = MEPop/MEBeds

VTdata = icudata[icudata['State'] == 'Vermont']
VTPop = sum(VTdata['Total Population'])
VTBeds = sum(VTdata['ICU Beds'])
VTPopBed = VTPop/VTBeds

MAdata = icudata[icudata['State'] == 'Massachusetts']
MAPop = sum(MAdata['Total Population'])
MABeds = sum(MAdata['ICU Beds'])
MAPopBed = MAPop/MABeds

RIdata = icudata[icudata['State'] == 'Rhode Island']
RIPop = sum(RIdata['Total Population'])
RIBeds = sum(RIdata['ICU Beds'])
RIPopBed = RIPop/RIBeds

CTdata = icudata[icudata['State'] == 'Connecticut']
CTPop = sum(CTdata['Total Population'])
CTBeds = sum(CTdata['ICU Beds'])
CTPopBed = CTPop/CTBeds

NYdata = icudata[icudata['State'] == 'New York']
NYPop = sum(NYdata['Total Population'])
NYBeds = sum(NYdata['ICU Beds'])
NYPopBed = NYPop/NYBeds

NJdata = icudata[icudata['State'] == 'New Jersey']
NJPop = sum(NJdata['Total Population'])
NJBeds = sum(NJdata['ICU Beds'])
NJPopBed = NJPop/NJBeds

PAdata = icudata[icudata['State'] == 'Pennsylvania']
PAPop = sum(PAdata['Total Population'])
PABeds = sum(PAdata['ICU Beds'])
PAPopBed = PAPop/PABeds
```

Next, we can make a list of this data and save it to the name of the abbreviation for each state. Here, we have titled it *Northeastdata.*

The objective here is to make a bar graph with the x-axis being labels of the states. 

To do this, we can set the 1st entry of *Northeastdata* as *labels* and the 2nd entry as what will be our y-axis values. Then, we set our *xaxis* to be the labels of *Northeastdata.*

Next, we use *plt.bar* to make a bargraph of our data and *plt.ylim* to limit the y-axis to the interval of points our data falls in. 

Following this, we use *plt.xticks* to replace the default x-ticks with *xaxis* and then replace *xaxis* with *labels.*

Finally, we can put graph and axis titles using *plt.title(), plt.xlabel(),* and *plt.ylabel.* 

```python
Northeastdata = [('ME',MEPopBed),('NH', NHPopBed),('VT',VTPopBed),('MA',MAPopBed),('RI',RIPopBed),('CT',CTPopBed),('NY',NYPopBed),('NJ',NJPopBed),('PA',PAPopBed)]
print(Northeastdata)

labels, yaxis = zip(*Northeastdata)
xaxis = np.arange(len(labels)) 

plt.bar(xaxis, yaxis, width = 0.5, align='center', color = 'cornflowerblue')
plt.ylim(3500,7000)

plt.xticks(xaxis, labels) #Replace default x-ticks with xs, then replace xs with labels
plt.yticks(yaxis)

plt.title('Number of People per ICU Bed in US Northeast States')
plt.xlabel('Northeast States')
plt.ylabel('Number of People per ICU Bed')
```

### Conclusions Drawn From Graph
* Rhode Island has the least amount of people theoretically allocated to each ICU bed. In other words, there are more ICU Beds for its population than in other Northeast States. 
* Vermont has the most amount of people theoretically allocated to each ICU bed. 

# Plotting Marathon Data 

First, lets insert the marathon data from our machine. 

```python
marathondata = pd.read_csv(r"C:\Users\guptaj\Documents\marathon-data.csv")
```

Let's plot a histogram comparing the ages of male and female marathon runners! Now that we have loaded the data, we can save the data for ages of males in *male_ages* and ages for females in *female_ages.* 

```python
male_ages = marathondata[marathondata['gender'] == 'M']['age']
female_ages = marathondata[marathondata['gender'] == 'W']['age']
```

Now, we can plot two histograms, one for males and one for female, on one graph using the *.plot.hist.* Putting the two histograms on the same graph will allow for an easy comparison. We can also add a grid, adjust bin size and rwidth for visualization that is easy for the eyes to see. Also, we must make the histogram of each age a different color so that we can easily distinguish between the graphs. Then we can use *plt.legend()* to get a legend. 

Next, we can put graph and axis titles using *plt.title(), plt.xlabel(),* and *plt.ylabel.* 

Finally, we can add two lines that go through the median male age and the median female age to extract more information from this graph.

```python
male_ages.plot.hist(grid = True,bins=20, rwidth=0.7, color='darkturquoise', label='Male')
female_ages.plot.hist(grid = True,bins=20, rwidth=0.7, color='salmon', label='Female')
plt.legend()
plt.title('Histogram of Ages')
plt.xlabel('Ages')
plt.ylabel('Frequency')

plt.axvline(male_ages.median(), color='lightseagreen', linestyle='dashed', linewidth=3)
plt.axvline(female_ages.median(), color='tomato', linestyle='dashed', linewidth=3)
```

### Conclusions Drawn from the Graph

* The mean line shows male marathon runners are generally older than female marathon runners, since the median female age is within the lower 50% of male age data (to the left of the median male age line). 
* Since the frequency of male ages is consistently higher than females, there are more male marathon runners.
