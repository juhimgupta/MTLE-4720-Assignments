# Using Sympy to Create a Equation and Perform Basic Operations

## Part 1: Getting a Symbolic Equation 

__Step 1:__ First, we need to import *sympy* as *sp.*

```python
import sympy as sp 
```

__Step 2:__ Next, we can set C1,C2,C3,gamma1,gamma2, and gamma3 as symbols of those values in sympy. In order words, when we type in the left variables, sympy will be able to read them as those variables written as a string in red. This will help us to create a symbolic object using *sp.symbols.*

```python
C1,C2,C3,gamma1, gamma2, gamma3, x = sp.symbols('C1 C2 C3 \gamma_1 \gamma_2 \gamma_3 x')
```

__Step 3:__ After this, we can create a formula titled *c_func.* We use *sp.cos* and *sp.sin* to call on the sympy functions for cosine and sine. We can then type *c_func* to see what the function looks like! 

```python
c_func = C1*sp.sin(gamma1*x) + C2*sp.cos(gamma2*x) + C3*sp.sin(gamma3*x)
```

```python
c_func
```

__Step 4:__ Now we will try to substitute a value of n x pi for the symbol gamma in the expression where n1=2, n2=4, n3=6. We can get pi using the sympy pi object *sp.pi.* Here, we have saved these substituted values into *newc_func.*

```python
newc_func = c_func.subs({gamma1:sp.pi*2, gamma2:sp.pi*4, gamma3:sp.pi*6})
```

## Part 2: Making a Function to Input Values, to Plot, and to Perform Calculus 

__Step 1:__ We now can make a function that prompts users to enter values for *C1, C2,* and *C3,* solves the equation, and then plots the result between 0 and pi. We can use the *def* to define a function *sub,* where the user inputs values for *c1,c2,* and *c3.* In order to define our function, we need to ensure that everything we defined outside of the function *sub* is also defined inside of the function. 

After this, we can add a new variable *a* that specifices that we can substitute the inputted *c1, c2* and *c3* values for *C1, C2,* and *C3* which are in the original symbolic equation. 

To plot, we can just plot *a,* and to limit it to 0 and pi using *xlim.* 

We can also integrate the equation using the *sympy.integrate* and differentiate the equation using *sympy.diff.* In calculus, we always need to integrate or differentiate with respect to some variable. In this case, we integrate and differntiate with respect to *x.* 

```python
def sub(c1,c2,c3):
    C1,C2,C3,gamma1, gamma2, gamma3, x = sp.symbols('C1 C2 C3 \gamma_1 \gamma_2 \gamma_3 x')
    c_func = C1*sp.sin(gamma1*x) + C2*sp.cos(gamma2*x) + C3*sp.sin(gamma3*x)
    newc_func = c_func.subs({gamma1:sp.pi*2, gamma2:sp.pi*4, gamma3:sp.pi*6})
    a = newc_func.subs({C1:c1,C2:c2,C3:c2})
    
    sp.plot(a, xlim = (0,sp.pi))
    
    integrate = sp.integrate(a,x)
    print(integrate)
    differential = sp.diff(a,x)
    print(differential)
    
sub(1,2,3)
```

## Part 3: Solving the Equations For Initial Conditions

__Step 1:__ First, let's say that the equation *c_func* is equal to 0 and we want to solve for *C3* in terms of other variables. We can use *sp.solve* to do this! *sp.solve* assumes the equation is equal to 0 without any other input. So, we can simply specify the equation and what we want to solve for. 

```python
sp.solve(c_func,C3)
```

__Step 2:__ Next, let's solve the equation when it is equal to 2 and for *C3.* To set the equation equal to 2, we use *sp.Eq* To solve the equation for *C3,* we can use the same code used in the previous step. 

```python
sp.Eq(c_func,2)
sp.solve(c_func,C3)
```
